```python
def minmax_scale(array):
    max = array.max()
    min = array.min()
    return (array - min) / (max - min)
```

```python
scale_cols = ['A_opt', 'B_opt', 'C_opt', 'm_opt', 'n_opt',]

for col in scale_cols:
    colname = col + ' scaled'
    dyndb[colname] = minmax_scale(dyndb[col])
```
